```python
import pandas as pd
```

```python
data5 = pd.read_csv('../csv/bank-additional-full.csv', sep=';')
print(data5.shape)
data5.columns
```

```python
from funpymodeling import status
status(data5)
```

### poutcome and previus has 86% of null or zero data
#### so I will drop them

```python
data4 = data5.drop(['poutcome', 'previous'], axis=1)
data4.columns
```

### see unique values un object columns

```python
object_columns = data4.select_dtypes(include=['object']).columns

for column in object_columns:
   unique_values = data4[column].unique()
   print(f"\nUnique values in {column}:\n {unique_values}\n")
```

## FIll N/A with value 'other'

```python
data3 = data4.copy()
object_columns = data3.select_dtypes(include=['object']).columns

for column in object_columns:
   data3[column].fillna('other', inplace=True)



for column in object_columns:
   unique_values = data3[column].unique()
   print(f"\nUnique values in {column}:\n {unique_values}\n")

```

## Duration has nan and zeros. I don't like to engrose de main value. So I will keep the values in zero days.
### If this csv was for a predictive model I would transform the column into categories. And make a category for nan & zero

```python
data2=data3.copy()
data2['duration'] = data2['duration'].fillna(0)
status(data2)
```

```python
data1 = data2.copy()
data1.rename(columns={'y': 'investor'}, inplace=True)
status(data1)
```

```python
data1.to_csv('bank-additional_cleaned.csv')
```
